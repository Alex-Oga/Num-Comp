```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
pic1 = plt.imread('Ohno1.png')
plt.imshow(pic1)
plt.figure()
pic2 = plt.imread('Ohno2.png')
plt.imshow(pic2)
plt.figure()
pic3 = plt.imread('Ohno3.png')
plt.imshow(pic3)
plt.figure()
```

```python
#Finding length and width of the pictures
numX, numY, color = pic1.shape
numX, numY, color2 = pic2.shape
numX, numY, color3 = pic3.shape
picNum = 3 #Number of pictures
# For loop, loops through the same point in all pictures.
# Makes a new picture based on the color that is most common in all pictures
def originalPicture(pic1, pic2, pic3, numX, numY, color):
  original = pic1
  for i in range(numX):
    for j in range(numY): # Goes through all y coordinates on x first
      modeColor = []
      total1 = 0
      total2 = 0
      total3 = 0
      # Finds the total of the rbg values in the respecive pictures, to find which one occurs the most
      for k in range(color):
        total1 += pic1[i, j][k]
        total2 += pic2[i, j][k]
        total3 += pic3[i, j][k]
      modeColor.append(total1)
      modeColor.append(total2)
      modeColor.append(total3)
      tempPic1 = pic1[i, j]
      tempPic2 = pic2[i, j]
      tempPic3 = pic3[i, j]
      # Finds the most occuring rbg value in the pictures and then appends the values.
      if max(set(modeColor), key = modeColor.count) == total1: 
        original[i, j] = tempPic1
      elif max(set(modeColor), key = modeColor.count) == total1:
        original[i, j] = tempPic2
      else:
        original[i, j] = tempPic3
  plt.imshow(original)
  plt.figure()
print(originalPicture(pic1, pic2, pic3, numX, numY, color))
```
